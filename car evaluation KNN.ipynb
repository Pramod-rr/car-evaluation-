```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('car.data')
df.head()
```

```python
headers = ['buying','maint','door','persons','lug_boot','safty','class']
df.columns = headers
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder as le
le = le()
```

```python
buying = le.fit_transform(list(df["buying"]))
maint = le.fit_transform(list(df['maint']))
door = le.fit_transform(list(df["door"]))
persons = le.fit_transform(list(df['persons']))
lug_boot = le.fit_transform(list(df["lug_boot"]))
safty = le.fit_transform(list(df['safty']))
cls = le.fit_transform(list(df["class"]))
```

```python
features = list(zip(buying,maint,door,persons,lug_boot,safty)) 
labels = list((cls))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(features,labels,test_size=0.2)
```

```python
predict = 'class'
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=9)
model.fit(X_train,y_train)
```

```python
accuracy = model.score(X_test,y_test)
accuracy
```

```python
prediction = model.predict(X_test)
```

```python
categories = ['unacc','acc','good','vgood']

for _ in range(len(prediction)):
    print(f"Predicted:  {categories[prediction[_]]}, Actual: {categories[y_test[_]]}")
    distance, indices = model.kneighbors([X_test[_]],5,True)
    print("Distance :",distance)
    print('Indices :', indices)
```
